```python
# 資料清理
heart_data = heart_data.drop(['DEATH_EVENT' ,'creatinine_phosphokinase' ,'ejection_fraction' ,'serum_creatinine' ,'serum_sodium' ,'time'], axis=1)

heart_data['age'] = heart_data['age'].fillna(heart_data['age'].mean())
heart_data = heart_data.dropna(axis=0)

le = LabelEncoder()
heart_data['sex'] = le.fit_transform(heart_data['sex'])

pca = PCA(n_components=2)
df_pca = pca.fit_transform(heart_data)
```

```python
wcs = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters= i, random_state= 42)
  kmeans.fit(df_pca)
  wcs.append(kmeans.inertia_)

plt.plot(range(1, 11), wcs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCS')
plt.show()
```

```python
# 將模型分群數調至2
kmeans = KMeans(n_clusters= 2, random_state= 42)

# 建立模型並預測
kms_predict = kmeans.fit_predict(df_pca)
```

```python
# 印出各群
plt.scatter(df_pca[kms_predict == 0, 0], df_pca[kms_predict == 0, 1], s = 50, c = 'red', label = 'UnSurvived')
plt.scatter(df_pca[kms_predict == 1, 0], df_pca[kms_predict == 1, 1], s = 50, c = 'blue', label = 'Survived')


# 印出群心
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 50, c = 'yellow', label = 'Centroids')

plt.legend()
plt.show()
```

```python
heart_data['DEATH_EVENT'] = kms_predict
heart_data.head()
```
